```python
harga_satuan = {'laptop':8000000,
              'tablet':4000000,
              'smartphone':3000000}

keranjang = {} # dict kosong

while True:
    # input barang
    while True:
        beli = input('masukkan barang') # laptop
        if beli in harga_satuan.keys():
            break
        else:
            print('anda memasukkan barang yang tidak ada')

    # input jumlah
    while True:
        jumlah = input('jumlah barang')
        if jumlah.isdigit():
            jumlah = int(jumlah)
            break
        else:
            print('anda memasukkan bukan angka')

    # total harga
    total_harga = harga_satuan[beli] * jumlah

    # menambahkan barang yg dibeli dan total harga ke keranjang
    keranjang[beli] = total_harga

    # memastikan beli lagi atau engga
    while True:
        beli_lagi = input('beli lagi? (y/n)') # hanya y atau n
        if beli_lagi == 'y' or beli_lagi == 'n':
            break
        else:
            print('anda memasukkan bukan y atau n')

    if beli_lagi == 'n':
        break

print('\n----keranjang----')
for key, value in keranjang.items():
    print(f'\t-barang {key} harga {value}')
summ = sum(keranjang.values())
print(f'total pembelian {summ} \n')

if summ > 10000000:
  summ = summ * 0.8
  print(f'total pembelian setelah diskon {summ}')
else:
  print(f'total pembelian {summ}')
```
