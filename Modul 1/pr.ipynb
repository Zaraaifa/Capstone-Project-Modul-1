```python
#SOAL NOMOR 5
nama = input("Masukkan nama tamu : ")
print(f"Nama          : {nama}")
lama = int(input("Masukkan lama menginap : "))
print(f"Lama Menginap : {lama}")
tipe_kamar = input("Masukkan tipe kamar (Standard/Deluxe/Suite) : ").lower()
print(f"Tipe Kamar    : {tipe_kamar}")

dict_tamu = {'Nama': nama, 'Lama Menginap': lama, 'Tipe Kamar': tipe_kamar}

harga_kamar = {'standard': 500000, 'deluxe': 1000000, 'suite': 2000000}

if tipe_kamar in harga_kamar:
    if lama < 3:
        total_harga = harga_kamar[tipe_kamar] * lama
        print(f"Total harga sewa kamar Anda: Rp{total_harga}")
    elif 3 <= lama <= 6:
        print(f"Harga sewa kamar sebelum diskon: {harga_kamar[tipe_kamar]*lama}")
        total_harga = harga_kamar[tipe_kamar] * lama - ((harga_kamar[tipe_kamar] * 0.1))
        print(f"Total harga sewa kamar Anda setelah diskon: Rp{total_harga}")
    elif lama > 6:
        if tipe_kamar == 'standard':
            print(f"Harga sewa kamar sebelum diskon: {harga_kamar[tipe_kamar]*lama}")
            total_harga = harga_kamar[tipe_kamar] * lama - ((harga_kamar[tipe_kamar] * 0.15))
            print(f"Total harga sewa kamar Anda setelah diskon: Rp{total_harga}")
        elif tipe_kamar == 'deluxe':
            print(f"Harga sewa kamar sebelum diskon: {harga_kamar[tipe_kamar]*lama}")
            total_harga = harga_kamar[tipe_kamar] * lama - ((harga_kamar[tipe_kamar] * 0.20))
            print(f"Total harga sewa kamar Anda setelah diskon: Rp{total_harga}")
        elif tipe_kamar == 'suite':
            print(f"Harga sewa kamar sebelum diskon: {harga_kamar[tipe_kamar]*lama}")
            total_harga = harga_kamar[tipe_kamar] * lama - ((harga_kamar[tipe_kamar] * 0.25))
            print(f"Total harga sewa kamar Anda setelah diskon: Rp{total_harga}")
else:
    print("Tipe kamar tidak tersedia.")
```
