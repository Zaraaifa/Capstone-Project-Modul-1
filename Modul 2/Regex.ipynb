```python
import regex as re #as adalah alias untuk menyingkat
```

REGULER EXPRESSION & CHARACTER SETS

```python
text = 'Warna favorit brai adalah grey dan gray dan grai'
print(re.findall(r'[bg]r[ea][iy]', text)) #re.findall(r'[bg]r[ea][iy]', text)

nama = 'nama saya adalah andi abdi aldi aldo anda'
print(re.findall(r'a[bnl][ld][aio]', nama)) #dalam [] isinya character set sebagai variasi pencocokan
```

CHARACTER RANGES
[A-Za-z] = huruf besar dan kecil
[0-9] = angka
[A-Za-z0-9] = huruf besar dan kecil dan angka
[A-Za-z0-9_] = huruf besar dan kecil dan angka dan underscore

```python
text = 'Warna favorit brai adalah grAy, gray, gr4y, gruy, 9rOy, graI, grei'
print(re.findall(r'gr[a-zA-Z0-9][iIy]', text))
#case sensitive jadi harus spesifik
#9rOy tidak ditemukan karena karakter pertama hanya 'g'
```

NEGATIVE CHARACTER SETS [^...]

```python
text = 'bug hug dug'
re.findall(r'[^b]ug', text)
```

BOUNDARIES (\b) and RAW (r) FUNCTION

```python
text = 'cat, catch, catty'
re.findall(r'\bcat\b', text) #\b untuk potong awal dan akhir, 'r' untuk mengganggap sbg string
```

SOAL LATIHAN

```python
#1 
Kalimat1 = "i want to have a cat, but maybe bat is okay. I want to call her caty can you help me?"
print(f'soal 1 = {re.findall(r'\b[cb]at\b', Kalimat1)}')

#2
Kalimat2 = "I  prefer to write with a pencil instead of a pen." 
print(f'soal 2 = {re.findall(r'\bpen\b', Kalimat2)}')

#3
Kalimat3 = "She likes to jog, but not log or hog."
print(f'soal 3 = {re.findall(r'[^l]og', Kalimat3)}')

#4 biasa
Kalimat4 = "Please provide the code in the format 5TH 555 or 9YZ 987."
print(f"soal 4 = {re.findall(r'[0-9][A-Z][A-Z] [0-9][0-9][0-9]', Kalimat4)}")
```

```python
"""
SHORTHAND CHARACTER
\d = [0-9]
\D = [^0-9]
\s = [ \t\n\r\f\v]
\S = [^ \t\n\r\f\v]
\w = [a-zA-Z0-9_]
\W = [^a-zA-Z0-9_]
\b = [^a-zA-Z0-9_]
\B = [a-zA-Z0-9_] 

NON QUANTIFIED REPETITION
* = 0 or more hingga tak terbatas {0,}
+ = 1 or more
? = 0 or 1
"""
```

```python
#4 shorthand
Kalimat4 = "Please provide the code in the format 5TH 555 or 9YZ 987."
print(f"soal 4 = {re.findall(r'\b\w{3} \d\w{2}\b', Kalimat4)}")
```

```python
import regex as re

#SOAL 1
Advance1 = "Banyak akun dalam game GTA5 seperti Quick12 Brown34 mencoba menyerang Jumped567 dan Lazy89 dengan brutal."
print(f'1. Jawaban Advance1 = {re.findall(r'\b\w{4,}\d+\b', Advance1)}')

#SOAL 2
Advance2 = "kode buku yang saya pinjam AA1008, AA1009, AA976 ,AB20" 
print(f'2. Jawaban Advance2 = {re.findall(r'\w+\d+', Advance2)}')

# SOAL 3
Advance3 = "Ini adalah contoh URL: http://www.example.com dan juga https://www.example2.com" 
print(f'3. Jawaban Advance3 = {re.findall(r'https?://\S+', Advance3)}')
print(f'3. Jawaban Advance3_2 = {re.findall(r'https?://www.\w+.\w+', Advance3)}')

#SOAL 4
Advance4 = "kode IP address 123.23.12.39.99 dan 142.11.12.27.55 dalam version 1.2.1.3" 
print(f'4. Jawaban Advance4 = {re.findall(r'\d+\.\d+\.\d+\.\d+.\d+', Advance4)}')

#SOAL 5
Advance5 = "Tanggal 1: 01/08/2023, Tanggal 2: 31-12-2022" 
print(f"5. Jawaban Advance5 = {re.findall(r'\d+[/-]\d+[/-]\d+', Advance5)}")

#SOAL 6
Advance6 = "Nomor plat: B 1234 CD, B 3876 RFE, AB 2394 AP, RI 1"
print(f"6. Jawaban Advance6 = {re.findall(r'\w+ \d+ ?\w*', Advance6)}")

#SOAL 7
Advance7 = "Temukan saya di Instagram @john_doe1 , @janesmith5 dan @abdillah.zaraifa"
print(f"7. Jawaban Advance7 = {re.findall(r'@\S+', Advance7)}")

#SOAL 8
Advance8 = "email saya fathindifa@purwadhika.com, fathin_difa@gmail.com, fathin26@yahoo.co.id"
print(f"8. Jawaban Advance8 = {re.findall(r'\b\w+@\S*\b', Advance8)}")
```

GROUPING

```python
Advance4 = "kode IP address 123.23.12.39.99 dan 142.11.12.27.55 dalam version 1.2.1.3" 
re.findall(r'(?:\d+.){5}', Advance4)
```

```python
with open('mbox-short.txt','r') as infile:
    text = infile.read()
```

```python
#QUESTION 1
email = re.findall(r'\b\w+@\w*\.\S*\b', text)
for i in email:
    print(i)

print(f"\njumlah email = {len(email)}")
```

```python
#QUESTION 2
date = re.findall(r'\b\d{4}-\d{1,2}-\d{1,2} \d{1,2}\:\d{1,2}\:\d{1,2}\b', text)
for i in date:
    print(i)

print(f"\njumlah tanggal = {len(date)}")
```

```python
#QUESTION 3
ip_address = re.findall(r'\b(?:\d+\.){3}\d{1,3}\b', text)
for i in ip_address:
    print(i)

print(f"\njumlah IP address = {len(ip_address)}")
```

```python
#QUESTION 4
day = re.findall(r'\b\w{3}, \d+ \w{3} \d+\b', text)
for i in day:
    print(i)

print(f"\njumlah tanggal = {len(day)}")
print(f"""jumlah per hari:
- Monday = {len(re.findall(r'\bMon, \d+ \w{3} \d+\b', text))}
- Tuesday = {len(re.findall(r'\bTue, \d+ \w{3} \d+\b', text))}
- Wednesday = {len(re.findall(r'\bWed, \d+ \w{3} \d+\b', text))}
- Thursday = {len(re.findall(r'\bThu, \d+ \w{3} \d+\b', text))}
- Friday = {len(re.findall(r'\bFri, \d+ \w{3} \d+\b', text))}
- Saturday = {len(re.findall(r'\bSat, \d+ \w{3} \d+\b', text))}
- Sunday = {len(re.findall(r'\bSun, \d+ \w{3} \d+\b', text))}
""")
```
